```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset simulating major tech layoffs from 2023 to 2025
data = {
    "Company": ["Intel", "Google", "Meta", "Amazon", "Microsoft"],
    "Date": pd.to_datetime(["2025-04-23", "2024-01-15", "2023-11-09", "2024-03-27", "2023-09-30"]),
    "Employees_Laid_Off": [25000, 12000, 11000, 9000, 10000],
    "Total_Workforce": [125000, 190000, 87000, 1500000, 221000],
    "Region": ["Global", "Global", "US/EU", "Global", "Global"]
}

# Create DataFrame
df = pd.DataFrame(data)
df["Percent_Laid_Off"] = (df["Employees_Laid_Off"] / df["Total_Workforce"]) * 100

# Sort by date for plotting
df_sorted = df.sort_values("Date")

# Plot: Percent of Workforce Laid Off Over Time
plt.figure(figsize=(10, 6))
plt.plot(df_sorted["Date"], df_sorted["Percent_Laid_Off"], marker='o')
for i, txt in enumerate(df_sorted["Company"]):
    plt.annotate(txt, (df_sorted["Date"].iloc[i], df_sorted["Percent_Laid_Off"].iloc[i]), textcoords="offset points", xytext=(0,10), ha='center')
plt.title("Tech Company Layoffs (2023–2025)")
plt.xlabel("Date")
plt.ylabel("% of Workforce Laid Off")
plt.grid(True)
plt.tight_layout()
plt.savefig("my_plot.png", dpi=300)
plt.show()
```
